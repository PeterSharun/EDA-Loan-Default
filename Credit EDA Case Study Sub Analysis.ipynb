## Step 2: Data Cleaning

-  ### Substep 2.1: Inspect the null values

```python
# Write your code for column-wise null count here
print("Column-wise null count and percentage:")

# Calculating column-wise null count
col_null_count_series = previous_application_df.isnull().sum()

# Calculating column-wise null percentage
col_null_perc_series = round(((col_null_count_series / previous_application_df.shape[0]) * 100), 2)

# Combining column-wise null count and null percentages
combined_col_null_df = pd.concat([col_null_count_series, col_null_perc_series], axis=1)

# Changing column names
combined_col_null_df.columns = ['Column-wise Null Count', 'Column-wise Null Percentage']

# Printing the combined dataframe
display(combined_col_null_df)
```

```python
# Write your code for row-wise null count here
print("Row-wise null count and null percentage:")

# Calculating row-wise null count
row_null_count_series = previous_application_df.isnull().sum(axis=1)

# Calculating column-wise null percentage
row_null_perc_series = round(((row_null_count_series / previous_application_df.shape[1]) * 100), 2)

# Combining column-wise null count and null percentages
combined_row_null_df = pd.concat([row_null_count_series, row_null_perc_series], axis=1)

# Changing column names
combined_row_null_df.columns = ['Row-wise Null Count', 'Row-wise Null Percentage']

display(combined_row_null_df)
```

-  ### Substep 2.2: Drop unecessary columns
    2.2.1: Dropping coloumns whose null percentages is very high
    
    2.2.2: Many of the columns are of less significance so it is better to drop those unecessary columns, they are:
    
    
        - WEEKDAY_APPR_PROCESS_START
        - HOUR_APPR_PROCESS_START
        - FLAG_LAST_APPL_PER_CONTRACT
        - NFLAG_LAST_APPL_IN_DAY
        - RATE_DOWN_PAYMENT
        - NAME_PAYMENT_TYPE
        - NAME_TYPE_SUITE
        - SELLERPLACE_AREA
        - NAME_SELLER_INDUSTRY
        - CNT_PAYMENT
        - PRODUCT_COMBINATION
        - DAYS_FIRST_DRAWING
        - DAYS_FIRST_DUE
        - DAYS_LAST_DUE_1ST_VERSION
        - DAYS_LAST_DUE
        - DAYS_TERMINATION

```python
# 2.2.1

# Calculating whether null % is greater than 45%
combined_col_null_df['Is Greater than 45%'] = combined_col_null_df['Column-wise Null Percentage'] > 45

# Printing the combined dataframe
print("Null percentage of columns:")
display(combined_col_null_df)

# Calculating the columns for which null % is greater than 45%
null_percen_cols_great5_index = combined_col_null_df[combined_col_null_df['Is Greater than 45%']].index

print("\n---------------------------------------------------------------------------------------------------\n")

# Printing columns for which null % is greater than 45%
print("Columns for which null percentage is greater than 45% :")
print("- Number of columns: {0}".format(len(list(null_percen_cols_great5_index))))
print("- List of columns: {0}".format(list(null_percen_cols_great5_index)))

print("\n---------------------------------------------------------------------------------------------------\n")

# Removing columns for which null percentage is greater than 45%
print("Dropping columns for which null percentage is greater than 45%...")
previous_application_df_coldropv1 = previous_application_df.drop(columns=null_percen_cols_great5_index, axis=1)

print("\n---------------------------------------------------------------------------------------------------\n")

# Calculating column-wise null percentage after dropping
col_null_count_series_coldrop = previous_application_df_coldropv1.isnull().sum()
col_null_perc_series_coldrop = round(((col_null_count_series_coldrop / previous_application_df_coldropv1.shape[0]) * 100), 2)
display(col_null_perc_series_coldrop)
print(previous_application_df_coldropv1.shape)
```

```python
# 2.2.2

# Code for dropping the unecessary columns

# Printing number of columns before dropping
print("Number of columns before dropping : {0} \n".format(previous_application_df_coldropv1.shape[1]))

#Preparing and printing the list of columns to be dropped
cols_to_be_dropped = [
'WEEKDAY_APPR_PROCESS_START',
'HOUR_APPR_PROCESS_START',
'FLAG_LAST_APPL_PER_CONTRACT',
'NFLAG_LAST_APPL_IN_DAY',
# 'RATE_DOWN_PAYMENT',
'NAME_PAYMENT_TYPE',
# 'NAME_TYPE_SUITE',
'SELLERPLACE_AREA',
'NAME_SELLER_INDUSTRY',
'CNT_PAYMENT',
'PRODUCT_COMBINATION',
'DAYS_FIRST_DRAWING',
'DAYS_FIRST_DUE',
'DAYS_LAST_DUE_1ST_VERSION',
'DAYS_LAST_DUE',
'DAYS_TERMINATION'
]
print("Number of columns to be dropped : {0} \n \nList of columns to be dropped : {1}\n".format(len(cols_to_be_dropped), cols_to_be_dropped))

# Dropping olumnns and assigning it to new dataframe "movies_after_coldrop"
print("Dropping columns...\n")
prev_app_after_coldrop = previous_application_df_coldropv1.drop(columns=cols_to_be_dropped, axis=1)

# Printing number of columns after dropping
print("Number of columns after dropping : {0}\n".format(prev_app_after_coldrop.shape[1]))

# Printing the columns of resultant dataframe
print("Resultant columns of dataframe : {0}".format(list(prev_app_after_coldrop.columns)))

# Printing the head of resultant dataframe
print("\n---------------------------------------------------------------------------------------------------\n")

print("Resultant dataframe (first 5 rows) :")

display(prev_app_after_coldrop.head(5))

print("\n---------------------------------------------------------------------------------------------------\n")

# Calculating column-wise null percentage after dropping
col_null_count_series_coldrop = prev_app_after_coldrop.isnull().sum()
col_null_perc_series_coldrop = round(((col_null_count_series_coldrop / prev_app_after_coldrop.shape[0]) * 100), 2)
display(col_null_perc_series_coldrop)
print(prev_app_after_coldrop.shape)
```

-  ### Substep 2.3: Drop unecessary rows using columns with high Null percentages

```python
# Code for dropping the rows here

# Calculating null percentages of columns
null_percen_cols = round((((prev_app_after_coldrop.isnull().sum() / prev_app_after_coldrop.shape[0]) * 100)),2)
print("->Null percentages of columns : \n\n{0}".format(null_percen_cols))

print("\n---------------------------------------------------------------------------------------------------\n")

print("Note:")
print("- We can see that there are three columns with null values.")
print("- NFLAG_INSURED_ON_APPROVAL has 40.3% of null values, if we remove the rows based on this column then there will too many dropped rows which might affect our analysis.")
print("- So, if we drop the rows for the remaining two columns ('AMT_ANNUITY' & 'AMT_GOODS_PRICE') which has fewer null values, we will have good retainment ratio.")
print("- Also this will reduce the number of null values in the 'NFLAG_INSURED_ON_APPROVAL' column which can be handled by imputing values.")

print("\n---------------------------------------------------------------------------------------------------\n")

# Finding columns for which null percent is greater than 5%
cols_to_remove = ['AMT_ANNUITY', 'AMT_GOODS_PRICE']

# Removing rows which have null values for columns whose null percentage is greater than 5%
print("Removing rows which have null values for columns 'AMT_ANNUITY' and 'AMT_GOODS_PRICE'...")
prev_app_after_rowcoldrop = prev_app_after_coldrop.dropna(axis=0, subset=cols_to_remove)

print("\n---------------------------------------------------------------------------------------------------\n")

# Calculating null percentages of columns after dropping rows
null_percen_cols_aftDrop = round((((prev_app_after_rowcoldrop.isnull().sum() / prev_app_after_rowcoldrop.shape[0]) * 100)), 2)
print("->Null percentages of columns after dropping rows : \n\n{0}".format(null_percen_cols_aftDrop))

print(prev_app_after_rowcoldrop.shape)
```

-  ### Substep 2.4: Fill NaN values

    We can see that there are null values only in 'NFLAG_INSURED_ON_APPROVAL' column which we will resolve by imputing data.

```python
# Imputing values for null values in 'NFLAG_INSURED_ON_APPROVAL' column

print("Frequency of values in 'NFLAG_INSURED_ON_APPROVAL' column before imputing:")
print(prev_app_after_rowcoldrop['NFLAG_INSURED_ON_APPROVAL'].value_counts())

print("\n---------------------------------------------------------------------------------------------------\n")

print("We can see that most of the client did not request for insurance during the previous application")
print("Therefore, we can replace the missing values with the mode.")
print("Imputing missing values...")

print("\n---------------------------------------------------------------------------------------------------\n")

mode_insurance = prev_app_after_rowcoldrop['NFLAG_INSURED_ON_APPROVAL'].mode()[0]
prev_app_after_rowcoldrop['NFLAG_INSURED_ON_APPROVAL'].fillna(mode_insurance, inplace=True)
print("Frequency of values in 'NFLAG_INSURED_ON_APPROVAL' column after imputing :")
display(prev_app_after_rowcoldrop['NFLAG_INSURED_ON_APPROVAL'].value_counts())
```

-  ### Substep 2.5: Drop duplicate values (if any)

```python
# Write your code for dropping duplicate values here

# Calculating number of rows before dropping
print("Number of rows before dropping : {0}\n".format(prev_app_after_rowcoldrop.shape[0]))

# Dropping duplicates
print('Dropping duplicates... \n')
movies_after_colrowdrop_dropdup = prev_app_after_rowcoldrop.drop_duplicates(keep = 'first')

# Calculating number of rows after dropping
print("Number of rows after dropping : {0}".format(movies_after_colrowdrop_dropdup.shape[0]))

print('\nWe can observe that there are no duplicate values.')
```

-  ### Subtask 2.6: Check the number of retained rows

```python
# Write your code for checking number of retained rows here

# Calculating null percentage of columns
print("Null percentage of columns : \n{0}".format(round(((prev_app_after_rowcoldrop.isnull().sum() / prev_app_after_rowcoldrop.shape[0]) * 100), 2)))

print("\n---------------------------------------------------------------------------------------------------\n")

# Calculating number and percentage of rows retained
print("Number and percentage of rows retained after completing all the tasks above :")
print("Number of rows before dropping : {0}".format(previous_application_df.shape[0]))
print("Number of rows retained after dropping : {0}".format(prev_app_after_rowcoldrop.shape[0]))
retained_row_perc = (prev_app_after_rowcoldrop.shape[0] / previous_application_df.shape[0]) * 100
print("Percentage of retained rows : {0}".format(retained_row_perc))
```
